```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_breast_cancer
import sklearn.metrics as m
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
# This is the data you need to create dataFrame with the given entries and save in Countryclusters.csv
col_names = ['Country','Latitude','Longitude','Language']
data= [['USA',44.97,-103.77,'English'],
       ['Canada',62.4,-96.8,'English'],
       ['France',46.75,2.4,'French'],
       ['UK',54.01,-2.53,'English'],
       ['Germany',51.15,10.4,'German'],
       ['Australia',-25.45,133.11,'English']]
df1 = pd.DataFrame(data, columns=col_names)
df1.head()
```

```python
# Load and visualize the data
df1.plot();
```

```python
#Visualize the scatter plot of Longitude and Latitude
plt.scatter('Longitude','Latitude', data=df1)
plt.show()
```

```python
 x = df1.iloc[:,1:3] # First for rows and second for columns
 x
```

```python
# train Kmeans model on x
kmeans = KMeans(2, n_init=10)
kmeans.fit(x)
```

```python
# predict on x
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
#Visualize the scatter plot of Longitude and Latitude again but this time with clusters identification
plt.scatter('Longitude','Latitude', data=df1, c=identified_clusters)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# perform an elbow method using multiple values of K on x-axis and WCSS (within cluster sum of square)
wcss = []
for i in range(1, df1.shape[0]+1):
    kmeans = KMeans(i, n_init=10, random_state=53)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
print(f"WCSS: {wcss}")

plt.plot(range(1, df1.shape[0]+1), wcss, marker='o')
plt.title("Elbow Method For Optimal K")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# view how many different types of variables are there in status_id
print(f"Number of different variables in 'status_id': {df['status_id'].nunique()}")

# view the labels in the variable
print(f"Labels of different variables in 'status_id': {df['status_id'].unique()}")

# view how many different types of variables are there in status_published
print(f"\n\nNumber of different variables in 'status_published': {df['status_published'].nunique()}")

# view the labels in the variable
print(f"Labels of different variables in 'status_published': {df['status_published'].unique()}")

# view how many different types of variables are there in status_type
print(f"\n\nNumber of different variables in 'status_type': {df['status_type'].nunique()}")

# view the labels in the variable
print(f"Labels of different variables in 'status_type': {df['status_type'].unique()}")
```

```python
# drop 'status_id', 'status_published'
df.drop(['status_id', 'status_published'], axis=1, inplace=True)
df.head()
```

```python
# declare features and target
features = df.drop('status_type', axis=1)
target = df['status_type']


# encode both features and labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
encoded_features = features.apply(label_encoder.fit_transform)
encoded_target = label_encoder.fit_transform(target)


# visualize
print(f"Encoded Target:\n{encoded_target}")

print("\n\nEncoded Features:\n")
encoded_features
```

```python
# perform normalization using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(encoded_features)

print(f"Normalized Features:\n{normalized_features}")
```

```python
# convert data into dataFrame
normalized_df = pd.DataFrame(normalized_features, columns=encoded_features.columns)
normalized_df.head()
```

```python
# perform KM clustering
km = KMeans(4, n_init=10, random_state=55)
clusters = km.fit_predict(normalized_df)
clusters
```

```python
# find the number of correctly label samples
correct_labels = sum(clusters == encoded_target)
print(f"Number of Correctly Labeled Samples: {correct_labels}")

# check how many of the samples were correctly labeled)
print(f'Accuracy score: {correct_labels/float(encoded_target.size):.2f}'.format())
```

```python
# Implement elbow method and plot the Within-Cluster-Sum-of-Squares (WCSS)
wcss1 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(normalized_df)
    wcss1.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss1, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster-Sum-of-Squares (WCSS)')
plt.show()
```
